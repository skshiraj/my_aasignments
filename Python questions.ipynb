             [3 4]]

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples


# you can free to change all these codes/structure
# here A and B are list of lists

# Program 1

def matrix_mul(A, B):
    
    R = []
    
    # For A
    no_row_A = len(A)
    no_col_A = len(A[0])

    # For B
    no_row_B = len(B)
    no_col_B = len(B[0])

    if no_col_A == no_row_B:# For A*B
        for i in range(no_row_A):
            l1 = []
            for j in range(no_col_B):
                p = 0
                for k in range(no_row_B):
                    p = p + A[i][k] * B[k][j]
                l1.append(p)
            R.append(l1)
        return(R)
    else:
        print("Since no of columns of first matrix is not equal to the no of rows of second matrix,therefore multiplication is not possible")
        return 0       
    
A = [[1,5],
     [5,3],
     [0,2]]

B = [[1,2,3],
     [5,3,1]]

r = matrix_mul(A, B)
print(r)
```

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples

# you can free to change all these codes/structure
import numpy as np # importing numpy just for using arange() function
import math
def display_dash_board(students, marks):
    # write code for computing top top 5 students
    # mapping of two lists :https://www.w3resource.com/python-exercises/dictionary/python-data-type-dictionary-exercise-13.php
    keys = students
    values = marks
    dictionary = dict(zip(keys, values))
    # sorting of dictionary :https://www.saltycrane.com/blog/2007/09/how-to-sort-python-dictionary-by-keys/
    top_5_students = sorted(dictionary.items(), key=lambda item: item[1],reverse = True)[:5] # storing only first five values

    # write code for computing top least 5 students
    least_5_students = sorted(dictionary.items(), key=lambda item: item[1])[:5]
    
    # write code for computing top least 5 students
    marks.sort()
    l = [25,75]
    lst = []
    for i in l:
        p1 = (i/100) * (len(marks))
        inNumberint = int(p1)
        if p1 == inNumberint:
            p = int(p1)
            per = marks[p-1]
            lst.append(per)
        else:
            k1 = math.floor(p1)
            k2 = math.ceil(p1)
            per = (marks[k1-1] + marks[k2-1])/2
            lst.append(per)
    lst1 = []
    for i in dictionary.values():
        if i in np.arange(lst[0],lst[1]+1,0.5):
            lst1.append(i)
    # lst1 contains marks that lies between 25 and 75 percentile 

    m1 = [45, 78, 12, 14, 48, 43, 47, 98, 35, 80] # Taking original list of marks not sorted
    keys = m1
    values = students
    rev_dictionary = dict(zip(keys, values)) # creating new dictionary with marks as keys and students as values
    
    #https://stackoverflow.com/questions/44434575/select-the-keyvalue-pairs-from-a-dictionary-in-python
    students_within_25_and_75 =  {x: rev_dictionary[x] for x in lst1} 

            
    return top_5_students, least_5_students, students_within_25_and_75

students=['student1','student2','student3','student4','student5','student6','student7','student8','student9','student10'] 
marks = [45, 78, 12, 14, 48, 43, 47, 98, 35, 80]
top_5_students, least_5_students, students_within_25_and_75 = display_dash_board(students, marks)
print("top_5_students:",top_5_students)
print("least_5_students",least_5_students)
print("students_within_25_and_75:",students_within_25_and_75)
```

<h3> Q5: Find the closest points</h3>

Consider you are given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P

<br>Cosine distance between two points (x,y) and (p,q) is defined as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples
# you can free to change all these codes/structure


# here S is list of tuples and P is a tuple ot len=2
def closest_points_to_p(S, P):
    p1 = list(P)

    lst_d = []
    for i in S: 
        lst = []
        lst = list(i)
        q1 = lst[0]*p1[0] + lst[1]*p1[1]
        q2 = math.sqrt((lst[0] * lst[0]) + lst[1] * lst[1])
        q3 = math.sqrt((p1[0] * p1[0]) + p1[1] * p1[1])
        cos_d = math.acos(q1/(q2*q3))
        lst_d.append(cos_d)
    keys = S
    values = lst_d
    dictionary = dict(zip(keys, values))
    l = sorted(dictionary.items(), key=lambda item: item[1])[:5]
    closest_points_to_p1 = []
    for i in l:
        g = list(i)
        closest_points_to_p1.append(g[0])
    return closest_points_to_p1  

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
points = []
points = closest_points_to_p(S, P)
print(points)
```
